```python
from render_typ import FormulaRenderer
```

```python
formulas = [
    r'$ A = pi r^2 $',
    r'$ "area" = pi dot "radius"^2 $',
    r'$ cal(A) := { x in RR | x "is natural" } $',
    r'$ x < y => x gt.eq.not y $',
    r'$ mat(1, 2; 3, 4) $',
    r'''$ sum_(k=0)^n k
    &= 1 + ... + n \
    &= (n(n+1)) / 2 $''',
]
```

```python
basic_renderer = FormulaRenderer()

for f in formulas:
    basic_renderer.render(f)
```

```python
# Fixed-width renderer
fixed_width_renderer = FormulaRenderer(override_config="set page(width: 200pt)")

for f in formulas:
    fixed_width_renderer.render(f)
```

```python
# Customized background color
custom_bg_renderer = FormulaRenderer(override_config=r'set page(fill: gradient.linear(..color.map.crest))')

for f in formulas:
    custom_bg_renderer.render(f)
```

```python
named_formulas = {
    "mass_energy_equivalence": r'$ E = m c^2 $',
    "euler_identity": r'$ e^(i pi) + 1 = 0 $',
    "general_relativity": r'$ G_(mu nu) + Lambda g_(mu nu) = kappa T_(mu nu) $',
}

for k, v in named_formulas.items():
    basic_renderer.render(v, name=k)
```
